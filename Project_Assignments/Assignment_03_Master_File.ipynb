```python
# Importing the required Libraries.

import selenium 
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException ,ElementNotInteractableException
from selenium.webdriver.common.by import By
import re
import time
import requests
import urllib
import csv


#Lets first connect to the driver-
driver = webdriver.Chrome(r"chromedriver.exe")

# Opening the Required URL of "https://maps.google.com"
url = "https://www.flipkart.com"
driver.get(url)
```

```python
# Getting inside to the search bar,

search_bar = driver.find_element(By.XPATH,'//input[@class="_3704LK"]')
search_bar.click()

# Passing the Input , in to the search bar .

search_bar.send_keys('Oneplus Nord')
search_bar.click()


# Click  ,to the Search icon.
search_bar = driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]')
search_bar.click()
```

```python
# Creating the Empty List.

Brand_Name = []
Smartphone_name = []
Colour = []
RAM = []
Storage_ROM = []
Primary_Camera = []
Secondary_Camera = []
Display_Size = []
Battery_Capacity = []
Price = []
product_urls = []
```

```python
# Fetching ALL URLS.

# Empty list , to store urls.
product_urls = []

# Scrapping all producct urls.
urls = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')

for i in urls[:24]:
    product_urls.append(i.get_attribute('href'))                             #appending the urls, in product_urls lists.

print(len(product_urls))
```

```python
# for first page product


for url in product_urls[:24]:
    driver.get(url)
    time.sleep(2)


# for Brand name scraping.(Brand_Name)
    brand_tag = driver.find_elements(By.XPATH,'//span[@class="B_NuCI"]')
    for i in brand_tag[:]:
        brand=i.text
        Brand_Name.append(brand)

        
# for Smartphone name scraping.(Smartphone_name)
    
    smart_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in smart_tag[2:3]:
        Smartphone=i.text
        Smartphone_name.append(Smartphone)

        
# for COLOUR scraping.(Colour)
    
    colour_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in colour_tag[3:4]:
        clr=i.text
        Colour.append(clr)

    readmore= driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')
    readmore.click()
        
# for Primary memory _RAM scraping.(RAM)
    
    primarymemo_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in primarymemo_tag[15:16]:
        Pmemo=i.text
        RAM.append(Pmemo)
        
# for Secondary memory _ROM scraping.(Storage_ROM)
    secondarymemo_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in secondarymemo_tag [14:15]:
        Smemo=i.text
        Storage_ROM.append(Smemo)
        
# for Primary Camera _ scraping.(Primary_Camera)
    pricamera_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in pricamera_tag [17:18]:
        pcam=i.text
        Primary_Camera.append(pcam)
        
# for Secondary Camera _ scraping.(Secondary_Camera)
    secondarycam_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in secondarycam_tag [18:19]:
        Scam=i.text
        Secondary_Camera.append(Scam)
        
        
# for Display Size _ scraping.(Display_Size)
    displaysize_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in displaysize_tag [9:10]:
        disp=i.text
        Display_Size.append(disp)
        
# for Battery Capacity _ scraping.(Battery_Capacity)
    battery_tag = driver.find_elements(By.XPATH,'//td[@class="URwL2w col col-9-12"]')
    for i in battery_tag [21:22]:
        batcap=i.text
        Battery_Capacity.append(batcap)
        
        
# for PRICE _ scraping.(Price)
    price_tag = driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
    for i in price_tag [:]:
        prc=i.text
        Price.append(prc)
                   
            
print(len(Brand_Name),len(Smartphone_name),len(Colour),len(RAM),len(Storage_ROM),len(Primary_Camera),len(Secondary_Camera),len(Display_Size),len(Battery_Capacity),len(Price))
print(Brand_Name,"\n\n",Smartphone_name,"\n\n",Colour,"\n\n",RAM,"\n\n",Storage_ROM,"\n\n",Primary_Camera,"\n\n",Secondary_Camera,"\n\n",Display_Size,"\n\n",Battery_Capacity,"\n\n",Price)
```

```python
# Making the Data Frame for the above details .
df=pd.DataFrame({'Seq_No':range(1,25,1),'Brand':Brand_Name,'Smartphone_NAME':Smartphone_name,'Colour':Colour,'RAM':RAM,'Storage_ROM':Storage_ROM,'PRICE':Price,'Primary_Camera':Primary_Camera,'Secondary_Camera':Secondary_Camera,'Display_Size':Display_Size,'Battery_Capacity':Battery_Capacity})
df
```

```python
# Saving the Data to CSV format.

df.to_csv('raw_data.csv',index=False)
```
